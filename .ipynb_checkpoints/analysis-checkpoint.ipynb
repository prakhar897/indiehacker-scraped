```python
import pandas as pd
import numpy as np
import json
```

```python
# Assuming you have two JSON files named file1.json and file2.json
with open('data1.json', 'r') as f:
    data1 = json.load(f)

with open('./data2.json', 'r') as f:
    data2 = json.load(f)
    
with open('./data3.json', 'r') as f:
    data3 = json.load(f)

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)
```

```python
import os
os.listdir()
```

```python
merged_df = pd.concat([df1, df2, df3], ignore_index=True)
merged_df
```

```python
merged_df = merged_df.drop_duplicates()
merged_df = merged_df.reset_index(drop=True)
len(merged_df)
```

```python
sorted_df = merged_df.sort_values(by='link')
sorted_df.head(50)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Count the occurrences of each unique value
counts = merged_df['revenue_explanation'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Revenue Explanation')
plt.show()
```

```python
def convert_to_numeric(value):
    try:
        return int(value.replace('$', '').replace(',', ''))
    except (ValueError, AttributeError):
        return None
    
rev_sorted_df = merged_df.sort_values(by='revenue')

rev_sorted_df["revenue_int"] = rev_sorted_df['revenue'].apply(convert_to_numeric)
```

```python
print(len(rev_sorted_df[rev_sorted_df["revenue_int"] >= 0]))
print(len(rev_sorted_df[rev_sorted_df["revenue_int"] == 0]))
print(len(rev_sorted_df[rev_sorted_df["revenue_int"] > 0]))
print(len(rev_sorted_df[rev_sorted_df["revenue_int"] > 100]))
print(len(rev_sorted_df[rev_sorted_df["revenue_int"] > 1000]))
print(len(rev_sorted_df[rev_sorted_df["revenue_int"] > 10000]))
```
